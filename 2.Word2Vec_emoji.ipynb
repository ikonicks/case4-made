# Ğ¡Ğ¾Ğ·Ğ´Ğ°Ğ½Ğ¸Ğµ Ğ¸ Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¸Ğµ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸ Ğ²ĞµĞºÑ‚Ğ¾Ñ€Ğ½Ñ‹Ñ… Ğ¿Ñ€ĞµĞ´ÑÑ‚Ğ°Ğ²Ğ»ĞµĞ½Ğ¸Ğ¹

Ğ¡Ñ€ĞµĞ´ÑÑ‚Ğ²Ğ°Ğ¼Ğ¸ Ğ±Ğ¸Ğ±Ğ»Ğ¸Ğ¾Ñ‚ĞµĞºĞ¸ `gensim` ÑĞ¾Ğ·Ğ´Ğ°Ğ¹Ñ‚Ğµ Ğ¸ Ğ¾Ğ±ÑƒÑ‡Ğ¸Ñ‚Ğµ Word2Vec Ğ¼Ğ¾Ğ´ĞµĞ»ÑŒ Ğ½Ğ° ÑÑ‚Ğ¾Ğ¼ Ğ½Ğ°Ğ±Ğ¾Ñ€Ğµ ÑĞ»Ğ¾Ğ². Ğ Ğ°Ğ·Ğ¼ĞµÑ€ Ğ²ĞµĞºÑ‚Ğ¾Ñ€Ğ½Ñ‹Ñ… Ğ¿Ñ€ĞµĞ´ÑÑ‚Ğ°Ğ²Ğ»ĞµĞ½Ğ¸Ğ¹ - 300, Ğ¿Ñ€Ğ¾Ñ‡Ğ¸Ğµ Ğ°Ñ€Ğ³ÑƒĞ¼ĞµĞ½Ñ‚Ñ‹ Ğ²Ñ‹Ğ±Ğ¸Ñ€Ğ°Ğ¹Ñ‚Ğµ ÑĞ°Ğ¼Ğ¾ÑÑ‚Ğ¾ÑÑ‚ĞµĞ»ÑŒĞ½Ğ¾, Ñ ÑƒÑ‡ĞµÑ‚Ğ¾Ğ¼ Ğ¿Ğ¾ÑĞ»ĞµĞ´ÑƒÑÑ‰Ğ¸Ñ… Ñ‚Ñ€ĞµĞ±Ğ¾Ğ²Ğ°Ğ½Ğ¸Ğ¹.
```python
from gensim.models import Word2Vec
from time import time

# Ğ¡Ğ¾Ğ·Ğ´Ğ°Ğ½Ğ¸Ğµ Ğ¼Ğ¾Ğ´ĞµĞ»Ğ¸ Word2Vec Ñ ÑƒĞºĞ°Ğ·Ğ°Ğ½Ğ½Ñ‹Ğ¼Ğ¸ Ğ¿Ğ°Ñ€Ğ°Ğ¼ĞµÑ‚Ñ€Ğ°Ğ¼Ğ¸:
# sentences=None - Ğ¿Ğ¾ĞºĞ° Ğ½Ğµ Ğ¿ĞµÑ€ĞµĞ´Ğ°ĞµĞ¼ Ğ´Ğ°Ğ½Ğ½Ñ‹Ğµ Ğ´Ğ»Ñ Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¸Ñ
# vector_size=300 - Ñ€Ğ°Ğ·Ğ¼ĞµÑ€Ğ½Ğ¾ÑÑ‚ÑŒ Ğ²ĞµĞºÑ‚Ğ¾Ñ€Ğ½Ğ¾Ğ³Ğ¾ Ğ¿Ñ€Ğ¾ÑÑ‚Ñ€Ğ°Ğ½ÑÑ‚Ğ²Ğ° (300 Ğ¿Ñ€Ğ¸Ğ·Ğ½Ğ°ĞºĞ¾Ğ² Ñƒ ĞºĞ°Ğ¶Ğ´Ğ¾Ğ³Ğ¾ ÑĞ»Ğ¾Ğ²Ğ°)
# workers=4 - ĞºĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ Ğ¿Ğ¾Ñ‚Ğ¾ĞºĞ¾Ğ² Ğ´Ğ»Ñ Ğ¿Ğ°Ñ€Ğ°Ğ»Ğ»ĞµĞ»ÑŒĞ½Ğ¾Ğ¹ Ğ¾Ğ±Ñ€Ğ°Ğ±Ğ¾Ñ‚ĞºĞ¸
# min_count=10 - Ğ¸Ğ³Ğ½Ğ¾Ñ€Ğ¸Ñ€Ğ¾Ğ²Ğ°Ñ‚ÑŒ ÑĞ»Ğ¾Ğ²Ğ°, Ğ²ÑÑ‚Ñ€ĞµÑ‡Ğ°ÑÑ‰Ğ¸ĞµÑÑ Ñ€ĞµĞ¶Ğµ 10 Ñ€Ğ°Ğ·
# negative=5 - ĞºĞ¾Ğ»Ğ¸Ñ‡ĞµÑÑ‚Ğ²Ğ¾ "negative samples" Ğ´Ğ»Ñ negative sampling Ğ°Ğ»Ğ³Ğ¾Ñ€Ğ¸Ñ‚Ğ¼Ğ° (ÑƒĞ»ÑƒÑ‡ÑˆĞ°ĞµÑ‚ ĞºĞ°Ñ‡ĞµÑÑ‚Ğ²Ğ¾)
# window=5 - Ğ¼Ğ°ĞºÑĞ¸Ğ¼Ğ°Ğ»ÑŒĞ½Ğ¾Ğµ Ñ€Ğ°ÑÑÑ‚Ğ¾ÑĞ½Ğ¸Ğµ Ğ¼ĞµĞ¶Ğ´Ñƒ Ñ‚ĞµĞºÑƒÑ‰Ğ¸Ğ¼ Ğ¸ Ğ¿Ñ€ĞµĞ´ÑĞºĞ°Ğ·Ñ‹Ğ²Ğ°ĞµĞ¼Ñ‹Ğ¼ ÑĞ»Ğ¾Ğ²Ğ¾Ğ¼ Ğ² ĞºĞ¾Ğ½Ñ‚ĞµĞºÑÑ‚Ğµ
model = Word2Vec(sentences=None, vector_size=300, workers=4, min_count=10, negative=5, window=5)

# Ğ—Ğ°ÑĞµĞºĞ°ĞµĞ¼ Ğ²Ñ€ĞµĞ¼Ñ Ğ½Ğ°Ñ‡Ğ°Ğ»Ğ° Ğ¿Ğ¾ÑÑ‚Ñ€Ğ¾ĞµĞ½Ğ¸Ñ ÑĞ»Ğ¾Ğ²Ğ°Ñ€Ñ
t = time()

# ĞŸĞ¾ÑÑ‚Ñ€Ğ¾ĞµĞ½Ğ¸Ğµ ÑĞ»Ğ¾Ğ²Ğ°Ñ€Ñ Ğ½Ğ° Ğ¾ÑĞ½Ğ¾Ğ²Ğµ Ñ‚Ñ€ĞµĞ½Ğ¸Ñ€Ğ¾Ğ²Ğ¾Ñ‡Ğ½Ñ‹Ñ… Ğ´Ğ°Ğ½Ğ½Ñ‹Ñ…:
# train_data - ÑĞ¿Ğ¸ÑĞ¾Ğº Ğ¿Ñ€ĞµĞ´Ğ»Ğ¾Ğ¶ĞµĞ½Ğ¸Ğ¹ (Ñ€Ğ°Ğ·Ğ±Ğ¸Ñ‚Ñ‹Ñ… Ğ¿Ğ¾ ÑĞ»Ğ¾Ğ²Ğ°Ğ¼) Ğ´Ğ»Ñ Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ¸Ñ
# progress_per=10000 - Ğ²Ñ‹Ğ²Ğ¾Ğ´Ğ¸Ñ‚ÑŒ ÑĞ¾Ğ¾Ğ±Ñ‰ĞµĞ½Ğ¸Ğµ Ğ¾ Ğ¿Ñ€Ğ¾Ğ³Ñ€ĞµÑÑĞµ Ğ¿Ğ¾ÑĞ»Ğµ ĞºĞ°Ğ¶Ğ´Ñ‹Ñ… 10000 Ğ¿Ñ€ĞµĞ´Ğ»Ğ¾Ğ¶ĞµĞ½Ğ¸Ğ¹
model.build_vocab(train_data, progress_per=10000)

# Ğ’Ñ‹Ğ²Ğ¾Ğ´ Ğ²Ñ€ĞµĞ¼ĞµĞ½Ğ¸, Ğ·Ğ°Ñ‚Ñ€Ğ°Ñ‡ĞµĞ½Ğ½Ğ¾Ğ³Ğ¾ Ğ½Ğ° Ğ¿Ğ¾ÑÑ‚Ñ€Ğ¾ĞµĞ½Ğ¸Ğµ ÑĞ»Ğ¾Ğ²Ğ°Ñ€Ñ (Ğ² Ğ¼Ğ¸Ğ½ÑƒÑ‚Ğ°Ñ…, Ñ Ğ¾ĞºÑ€ÑƒĞ³Ğ»ĞµĞ½Ğ¸ĞµĞ¼ Ğ´Ğ¾ 2 Ğ·Ğ½Ğ°ĞºĞ¾Ğ²)
print('Time to build vocab: {} mins'.format(round((time() - t) / 60, 2)))
```

```python
# Ğ’Ñ‹Ğ²ĞµÑÑ‚Ğ¸ ÑĞ»Ğ¾Ğ²Ğ° Ñ Ğ¸Ñ… Ğ¸Ğ½Ğ´ĞµĞºÑĞ°Ğ¼Ğ¸
for word, idx in list(model.wv.key_to_index.items())[:25]:  # Ğ¿ĞµÑ€Ğ²Ñ‹Ğµ 25
    print(f"{idx}: {word}")
```

```python
# Ğ¾Ğ±ÑƒÑ‡Ğ°ĞµĞ¼ Ğ¼Ğ¾Ğ´ĞµĞ»ÑŒ ~ 10 Ğ¼Ğ¸Ğ½ÑƒÑ‚
t = time()
model.train(train_data, total_examples=model.corpus_count, epochs=20, report_delay=1) # Ğ¾Ğ±ÑƒÑ‡Ğ°ĞµĞ¼ ~ 10 Ğ¼Ğ¸Ğ½
print('Time to train the model: {} mins'.format(round((time() - t) / 60, 2)))
```

*Ğ§ĞµĞº-Ğ¿Ğ¾Ğ¸Ğ½Ñ‚ 1: ĞœĞ¾Ğ´ĞµĞ»ÑŒ Ğ´Ğ¾Ğ»Ğ¶Ğ½Ğ° Ğ±Ñ‹Ñ‚ÑŒ Ğ¾Ğ±ÑƒÑ‡ĞµĞ½Ğ° Ñ‚Ğ°Ğº, Ñ‡Ñ‚Ğ¾Ğ±Ñ‹ ÑÑ€ĞµĞ´Ğ¸ Ğ¿Ğ¾Ñ…Ğ¾Ğ¶Ğ¸Ñ… Ñ‚Ğ¾ĞºĞµĞ½Ğ¾Ğ² ÑĞ¼Ğ¾Ñ‚Ğ¸ĞºĞ¾Ğ½Ğ¾Ğ² Ğ±Ñ‹Ğ»Ğ¾ Ğ½Ğµ Ğ¼ĞµĞ½ĞµĞµ 5.*

```python
def count_emoji(sentence):
  e_count=len([x[0] for x in sentence if is_emoji(x[0])]) # x[0] Ğ±ĞµÑ€ĞµÑ‚ Ğ¿ĞµÑ€Ğ²Ñ‹Ğ¹ ÑĞ¸Ğ¼Ğ²Ğ¾Ğ» ĞºĞ°Ğ¶Ğ´Ğ¾Ğ³Ğ¾ Ñ‚Ğ¾ĞºĞµĞ½Ğ° (ÑÑ‚Ğ¾ Ğ²Ğ°Ğ¶Ğ½Ğ¾ Ğ´Ğ»Ñ ÑĞ¼Ğ¾Ğ´Ğ·Ğ¸, Ñ‚Ğ°Ğº ĞºĞ°Ğº Ğ¾Ğ½Ğ¸ Ğ¾Ğ±Ñ‹Ñ‡Ğ½Ğ¾ Ğ·Ğ°Ğ½Ğ¸Ğ¼Ğ°ÑÑ‚ 1 ÑĞ¸Ğ¼Ğ²Ğ¾Ğ»)
  return e_count, len(sentence)-e_count
```
